## 🧪 Parte 5 — Teste de Hipóteses e P-Valor


**Template de exemplo (já feito para vocês):**
- **Hipótese:** o lucro médio (`Order Profit Per Order`) do *Standard Class* é diferente do *First Class*.
- **Teste:** Teste T de Student para amostras independentes.
- **Interpretação:** se *p-valor* < 0,05, rejeitamos a hipótese nula → a diferença é
  estatisticamente significativa (não é mero acaso).

```python
# === EXEMPLO RESOLVIDO: Teste T (Standard Class vs First Class) ===
standard = df.loc[df['Shipping Mode'] == 'Standard Class', 'Order Profit Per Order'].dropna()
first    = df.loc[df['Shipping Mode'] == 'First Class',    'Order Profit Per Order'].dropna()

t_stat, p_valor = stats.ttest_ind(standard, first, equal_var=False)  # Welch

print(f'Média Standard Class: {standard.mean():.2f}')
print(f'Média First Class:    {first.mean():.2f}')
print(f't = {t_stat:.3f} | p-valor = {p_valor:.4f}')
print('→', 'Rejeitamos H0: diferença significativa.' if p_valor < 0.05
      else 'Não rejeitamos H0: diferença pode ser acaso.')
```
